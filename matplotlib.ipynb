```python
import matplotlib.pyplot as plt 

x = [10, 20, 30, 40]
y = ["C", "C++", "Java", "Python"]
ex = [0.3, 0.0, 0.0, 0.0]
c = ["r", "b", "g", "y"]

plt.pie(x, labels=y, explode=ex, colors=c, autopct="%0.2f%%",shadow = True,radius = 1.3,labeldistance = 1.1,startangle = 0)

plt.show()
```

```python
import matplotlib.pyplot as plt       
x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [2,3,4,5,6]
area3 = [3,3,2,4,2]
l = ["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels = l, colors = ["r","g","b"])
plt.title("python")
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.show()
```
